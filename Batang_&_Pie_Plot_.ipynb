```python
#import dataset  ke dalam variabel dataframe
df_Sns6=pd.read_csv(filepath_or_buffer='Sns6.csv',sep=',')

 #preview data frame , 20 data  pertama (indeks 0-9)
df_Sns6.head(20)
```

```python
#buat obyek seaborn untuk graf  ini 
plt.figure(figsize =(15,5)) # set ukuran
Sns6_bar_plot = sns.barplot(x="tahun", y="indeks_pemberdayaan_gender",hue ="nama_kabupaten_kota", data= df_Sns6 )
Sns6_bar_plot.set_title('hasil Indeks Pemberdayaan Gender kabupaten/kota di provinsi jawa barat periode (2017-2018)')
```

```python
df_Sns5 = pd.read_csv('Sns5.csv', sep=',')

df_Sns5.head(30)
```

```python
plt.figure(figsize=(15, 15)) #set ukuran
plt.pie(df_Sns5['indeks_pembangunan_gender'] , labels=df_Sns5['nama_kabupaten_kota'], autopct ='%1.1f%%')
plt.title('Persentase pembangunan  gender  per kota atau kabupaten di provinsi jawa barat periode (2018-2020)')
plt.show()
```
