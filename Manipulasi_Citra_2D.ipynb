**Memotong Gambar (CROP)**

```python
img = Image.open("shrek.png")

img
```

```python
crop_img = img.crop((220, 20, 450, 220))

crop_img
```

**Membuat WaterMarks**

```python
# Watermark
from PIL import ImageDraw, ImageFont
```

```python
width , height = crop_img.size

draw = ImageDraw.Draw(crop_img)

text ='50419942'

font = ImageFont.truetype('arial.ttf', 18)

textwidth, textheight = draw.textsize(text, font)

margin = 15
x = width- textwidth- margin
y = height- textheight-margin

draw.text((x, y), text, font=font)
crop_img
```

**Rotate & Flip Gambar**

```python
#rotate gambar
rotated = crop_img.transpose(Image.ROTATE_90)

rotated
```

```python
#FLIP

flipped = crop_img.transpose(Image.FLIP_TOP_BOTTOM)

flipped
```

**#Brigtness, Contrass, dan Sharpnesss**

```python
from PIL import ImageEnhance
```

```python
#Brightness (kecerahan)

enhancer = ImageEnhance.Brightness(crop_img)

factor = 0.5 # Darker

im_output = enhancer.enhance(factor)

im_output
```

```python
#CONTRAST

enhancer = ImageEnhance.Contrast(crop_img)
factor = 1.5 # Add Contrast

im_output = enhancer.enhance(factor)

im_output
```

```python
#Sharpness
enhancer = ImageEnhance.Sharpness(crop_img)
factor = 1.5

im_output = enhancer.enhance(factor)

im_output
```
