# Membuat Scatter Plot Secara Programatis Menggunakan Python

Sumber data: 

Scatter plot adalah tipe plot atau diagram yang menggunakan koordinat Cartesian untuk menampilkan nilai data dua variabel dari sebuah dataset. Jika titik-titik pada scatter plot diberi kode (misalnya warna), maka satu variabel lagi bisa ditampilkan.

Umumnya, variabel independen ditampilkan di-plot sepanjang sumbu horizontal (sumbu x). Scatter plot dapat membantu kita mendeteksi korelasi antara dua variabel. Apabila titik-titik data membentuk pola dari kiri bawah sampai kanan atas diagram, maka mungkin ada korelasi positif. Jika pola menurun dari kiri atas sampai kanan bawah, maka korelasi negatif. (Soal ini tidak akan dibahas lebih lanjut; kita hanya fokus bagaimana membuat diagramnya.)

Plot akan dibuat menggunakan bahasa pemrograman Python, tepatnya modul [Seaborn](https://seaborn.pydata.org/index.html).

## Dataset Sumber

Sumber dataset: [Kaggle](https://www.kaggle.com/alexisbcook/scatter-plots/data?select=insurance.csv)

Unduh lalu upload di Colab via side menu Files, di lokasi yang sama dengan file notebook ini.

Di komputer lokal, buka file CSV di atas lalu amati datanya sekilas.

````python
# Lokasi file CSV
# Pastikan file nya benar-benar ada
insurance_filepath = "insurance.csv"

# Muat data dalam CSV ke dalam variabel dataframe pandas
insurance_data = pd.read_csv(insurance_filepath)

# Cek 10 data pertama
insurance_data.head(10)
````

## Menyiapkan Variabel Python dan Plotting

1. Buat variabel `x` dan `y`, masing-masing adalah kolom "body mass index (bmi)" dan "charges (biaya premi)"
2. Buat plot dengan fungsi `scatterplot()` dari modul `sns`

Dari plot yang dihasilkan, terlihat bahwa sepertinya orang dengan BMI tinggi harus membayar biaya premi asuransi kesehatan yang lebih tinggi.

````python
x = insurance_data['bmi']
y = insurance_data['charges']

# ada dua parameter wajib: x dan y
sns.scatterplot(x=x, y=y)
````

untuk memastikan kemungkinan korelasi, coba kita buat __garis regresi__ dengan fungsi `regplot()`

````python
# dua var ini sama dengan di atas
x = insurance_data['bmi']
y = insurance_data['charges']

sns.regplot(x=x, y=y)
````

## Menambahkan Kode Warna

Seperti sudah dijelaskan sebelumnya, jika kita melakukan pengkodean (dengan warna, misalnya), kita bisa menambahkan satu var lagi.

Coba kita tambahkan var "smoker (perokok)" sebagai var ketiga yang akan dikodekan dengan warna (hue)

Dari plot ini, sepertinya ada perbedaan biaya premi antara perokok dan non-perokok, walaupun BMI-nya mirip.

````python
# dua var ini sama dengan di atas
x = insurance_data['bmi']
y = insurance_data['charges']
hue = insurance_data['smoker']

sns.scatterplot(x=x, y=y, hue=hue)
````

Untuk membuktikan pernyataan terakhir di atas, mari kita coba membuat dua garis regresi sekaligus dengan fungsi `lmplot()`



````python
#sns.lmplot(x=x, y=x, hue=hue)
````

Kode 

````python
# ini kode yang benar
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
````
